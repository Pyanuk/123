Цель работы: заключалась в изучение визуализации данных в Python

```
import matplotlib.pyplot as plt
import pandas as pd


yandex = pd.read_csv("yandex_tracks_top100.csv")
yandex
```

```
plt.plot(yandex.genre)
plt.grid()
plt.show()
```

Данный линейный график нам показывает, что большинство слушаю кавказскую музыку(или их больше всего треков)

```
plt.plot(yandex["monthly_listens_total"].head(10), color="blue", label="Прослушивание за все время")
plt.ylabel("Количество прослушиваний")
plt.xlabel("Месяц")
plt.title('Прослушивание')
plt.grid()
plt.legend()
plt.show()
```

Данный линейный график показывеет прослушивание первых 20 треков, где видно, что за 1 месец было прослушано более 1.6ккк, а в начало 2 месеца всего лишь 6кк

```
plt.plot(yandex["artist(s)"].head(20), color = "blue", marker = "o", label="Артист")
plt.plot(yandex["genre"].head(20), color = "black", marker = "o", label="Жанр")
plt.legend()
plt.show()
```

Этот линейный график показывает артиста и его жанр.Например взять артиста MACAN его жанр песен русский рэп

```
plt.bar(yandex.index[:20], yandex["chart"].head(20), color="blue", label="Чарт")
plt.plot(yandex["artist(s)"].head(20), color = "blue", marker = "o", label="Артист")
plt.xlabel("Месяц")
plt.ylabel("Позиция в чарте")
plt.title('Чарт')
plt.legend()
plt.show()
```

Был создан столбчатый график чарта, где первые 20 песен в находятся в чарте, ANNA ASTI на первом месте, а XOLIDAYBOY находиться на 20.

```
plt.bar(yandex["name"].head(6), yandex["track_len"].head(6), color="pink", label="Продолжительность")
plt.ylabel('Продолжительность песни')
plt.title('Продолжительность')
plt.legend()
plt.show()
```

Данная диаграмма показывает на сколько идет песня первых 7 песен, видно что самая долгая песня длиться 2:48(А где прошла ты), а самая быстрая песня длить 1:54(Я плачу)

```
plt.bar(yandex["name"].head(5), yandex["artists_likes_total"].head(5), color="yellow")
plt.title('Лайки по песнями')

plt.show()
```

Эта диаграмма показывает что песня "ASPHALT" 8 набрал больше 2кк лайков, а песня "Я плачу" всего лишь +- 250к

```
content = yandex["Explicit_content"].value_counts().head()
plt.pie(content, labels = content.index, colors = ["blue", "red",],autopct="%1.1f%%")
plt.title("Соотношение явного контента")
plt.legend()
plt.show()
```

Данная круговая диаграмма показывает что явный контент только есть 22% артистов, а у остальных 78% он отсуствует

```
content = yandex["genre"].value_counts().head()
plt.pie(content, labels = content.index, colors = ["blue", "red","orange","pink","purple"],autopct="%1.1f%%")
plt.title("Соотношение жанра песен")
plt.legend()
plt.show()
```

Данна круговая диаграмма показывает что большинство артистов поют песни жанра русский реп 46.2%, а самый не популярный жанр среди артистов это "поп" который составляет 2.1%

```
plt.figure(figsize=(10, 6))
plt.scatter(yandex['artists_likes_total'].head(10), yandex['monthly_listens_total'].head(10), color='purple')
plt. title( 'Связь между лайком и прослушивание у первых 10' )
plt.xlabel('Лайки')
plt.ylabel('Просмотры')
plt.show()
```

Данная диаграмма рассеивания показывает отсуствие связи между лайками и просмотрами.

```
plt.figure(figsize=(10, 6))
plt.hist(yandex['genre'].head(30), bins=30, color=["pink"], edgecolor='orange',)
plt.xlabel('Жанр')
plt.ylabel('Сколько исполнителей пели')
plt.show()
```

Данная гистрограмма показывает что первые 30 артистов(большинсво), поют в жанре "русская поп-музыка", а самый не популярный жанр среди 30 артистов является "русский рок"

```
plt.figure(figsize=(10, 6))
plt.hist(yandex['Explicit_content'].head(20), bins=30, color=["pink"], edgecolor='orange',)
plt.ylabel('Кол-во явного контента')
plt.xlabel('Явный контен')
plt.show()
```

Данная гистрограмма показывает что среди исполнителей(30 человек) явный контент использует только 2.5 человек, а остальные его не используют
